```python
df5 = pd.DataFrame({
    'brand': ['Yum Yum', 'Yum Yum', 'Indomie', 'Indomie', 'Indomie'],
    'style': ['cup', 'cup', 'cup', 'pack', 'pack'],
    'rating': [4, 4, 3.5, 15, 5]
})
```

```python
df5
```

```python
df5.drop_duplicates()  # by default it will remove all duplicates according to rows of columns
```

```python
df5.drop_duplicates(subset=['brand'])  # simply it will remove duplicates from specific columns
```

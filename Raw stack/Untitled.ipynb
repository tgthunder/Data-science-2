```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
cv = CountVectorizer(max_features=7000,stop_words='english')
```

```python
vectors=cv.fit_transform(cars['feature']).toarray()
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity=cosine_similarity(vectors)
```

```python
def recommand():
    print(cars['company'].unique())
    obj1=input("Enter company:")
    print(cars[cars['company']==obj1]['Model name'].unique())
    obj2=input("Enter Model Name:")
    print(cars[cars['Model name']==obj2]['fuel'].unique())
    obj3=input("Enter Fuel Type:")
    print(cars[cars['Model name']==obj2]['engine'].unique())
    obj4=input("Enter engine:")
    car_index=cars[(cars['company']==obj1)&(cars['Model name']==obj2)&(cars['fuel']==obj3)&(cars['engine']==obj4)].index[0]
    k=similarity[car_index]
    lst=sorted(list(enumerate(k)),reverse=True,key=lambda x:x[1])[1:]
    lst1 = list(enumerate(cars['Insurance'].values.tolist()))
    lst2= []
    for i in lst:
        for j in lst1:
            if i[0]==j[0]:
                if j[1] not in lst2:
                    lst2.append(j[1])
  
    
    print("Your Insurance Plans")
    for i in lst2[0:5]:
        print(i)
    print("If you want more options I can Help You.Please Enter Y for More plans and N  for stop.")
    a = input("Enter Your choice:").lower()
    if a=='y':
        for i in lst2[5:10]:
            print(i)
    else:
        print("Thank you")
```

```python
recommand()
```
